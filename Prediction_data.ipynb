```python
# Import the pandas module
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("Final_Dataset.csv")

# Print the DataFrame
df.head()

# Get information about the DataFrame
df.info()

# Get descriptive statistics for the DataFrame
df.describe()
```

```python
categorical_cols = ['gender', 'residence','location' ]

# Performing one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols)

# Displaying the encoded DataFrame
print(df_encoded.head())
```

```python
df1=df_encoded
```

```python
X = df1.drop('TARGET_PREDICTION_PERCENT', axis=1)  # Features
y = df1['TARGET_PREDICTION_PERCENT']               # Target
```

```python
# Drop the non-numeric columns
X_train = X_train.drop(non_numeric_cols, axis=1)
X_test = X_test.drop(non_numeric_cols, axis=1)
```
